```python
#columns list
print(df.columns.tolist())
#for a particular column
print(df['Survived'].value_counts())  #which value has how many counts
print(df['Pclass'].unique())
#select multiple columns
print(df[['Pclass','Name','Age','Survived']].head(10))
```

Missing Values



```python
#how many nulls per column
print(df.isnull().sum())
df['Age']=df['Age'].fillna(df['Age'].median())
print(df['Age'].isnull().sum())
```

Filtering Rows

```python
survivors=df[df['Survived']==1]
print(survivors.shape)
```

```python
#first class passengers over age 40
fc_forty=df[(df['Age']>40) & (df['Pclass']==1)]
fc_forty.head()
print(fc_forty[['Name','Age','Pclass']].head())
```

```python
#survival rate
survival_by_class=df.groupby('Pclass')['Survived'].mean()
print(survival_by_class)
```

```python
#avg age per class
print(df.groupby('Pclass')['Age'].mean())
```
