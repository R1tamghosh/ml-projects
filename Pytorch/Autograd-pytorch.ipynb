Basically provides the automatic differentitation for tensor operations. thus it enables gradient calculations

```python
import torch
```

```python
x= torch.tensor(3.0,requires_grad=True)   #set to True if I want gradient at any point(makes a computation graph)
y=x**2
y
```

```python
y.backward()   #creates all the gradients in chain rule
```

```python
x.grad   #Not clear why need to do x.grad instead of y.grad(which is intuitive)
```

```python
x=torch.tensor(30.0,requires_grad=True) # Re-initialize x for a fresh graph, since each time I call backward function it consumes the computational graph
y=x**2 #Re-initialize y
z=torch.sin(y)
print(z)
print(y)
print(x)
z.backward() # This will now correctly compute dz/dx and store it in x.grad
x.grad
```

Applying this to do back propagation in a simple perceptron. With cross-entropy loss

```python
x=torch.tensor(7.4) #input
y=torch.tensor(0.0) #true-label
w=torch.tensor(1.0,requires_grad=True) #weight
b=torch.tensor(0.0,requires_grad=True) #bias
```

```python
#binary cross-entropy for scalar
def binary_cross_entropy(y_pred,y_true):
    epsilon=1e-8 #to prevent log(0)
    y_pred=torch.clamp(y_pred,epsilon,1-epsilon)  #keeping the predicted value in a given range
    return -(y_true*torch.log(y_pred)+(1-y_true)*torch.log(1-y_pred))
```

```python
#forward pass
z=w*x+b
y_pred=torch.sigmoid(z)
loss=binary_cross_entropy(y_pred,y)
print(loss)
```

```python
loss.backward()
print(w.grad)
print(b.grad)
```

```python
x1=torch.tensor([1.0,2.0,3.0],requires_grad=True)
y1=(x1**2).mean()
print(x1)
print(y1)
y1.backward()
x1.grad
```

Clearing gradient
