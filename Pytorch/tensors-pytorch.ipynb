Mathematical operations

```python
x2=torch.rand(3,3)
x2
```

```python
#addition
x2+2
#multi
x2*3
#subtraction
x2-1
#division
x2/2
#int_division
(x2*100)//3
#mod
((x2*100)//3)%2
#power
x2**2
```

Element wise operation


```python
#very similar to Numpy.(skipping)
c=torch.tensor([1,-2,3,4,-5])
#abs
print(torch.abs(c))
#negatve
print(torch.neg(c))
d=torch.tensor([1.9,2.3,4,5.4,6.7])
#round
print(torch.round(d))
#floor
print(torch.floor(d))
#ceil
print(torch.ceil(d))
#clamp-to keep the number in the tnsor in between a range
torch.clamp(d,min=2.2,max=5)
```

Reduction operations

```python
e=torch.randint(size=(4,5),low=0,high=10)
print(e)
torch.sum(e)
#colsum
print(torch.sum(e,dim=0))
#rowsum
print(torch.sum(e,dim=1))
#similar for mean,median,max,min,prod,std,var,argmax,argmin
print(torch.argmax(e))  #POSITION OF MAX ITEM, if used without dim then it flattens the tensor and then gives the item number
print(torch.argmin(e))
```

```python
#matrix mult
x3=torch.randint(size=(4,5),low=0,high=10)
x4=torch.randint(size=(5,4),low=0,high=10)
print(torch.matmul(x3,x4))
x5=torch.randint(size=(4,),low=0,high=10)
x6=torch.randint(size=(4,),low=0,high=10)
print(torch.dot(x6,x5))

#transpose
print(torch.transpose(x3,0,1))  #swap 0-th dim with 1-th dim
#similar with torch.det,torch.inverse
```

```python
#comparisons
x5>x5
#similar for other operations
```

Special functions

```python
x7=torch.randint(size=(4,5),low=0,high=10,dtype=torch.float32)
print(x7)
print(torch.log(x3))
print(torch.exp(x3))
print(torch.sqrt(x3))
print(torch.sigmoid(x3))
print(torch.softmax(x7,dim=0))
print(torch.relu(x7))
```
